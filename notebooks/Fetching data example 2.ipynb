# Fetching data via dataset ID

```python
import requests
import json
import yaml
import zipfile
import glob
```

```python
path = '../data/'
```

```python
headers = {
    'Content-Type': 'application/json',
}

json_data = {
  "query": {
    "datasets.dataset_id:any": [
      "8HoVaFVqTMSF60fQ8yyQpg"
    ]
  }
}

response = requests.post('http://localhost/nomad-oasis/api/v1/entries/raw/query', headers=headers, json=json_data)

with open(path + 'download.zip', 'wb') as f:
    f.write(response.content)
```

```python
response
```

```python
archive = zipfile.ZipFile(path + 'download.zip', 'r')
```

```python
archive.namelist()
```

```python
yamldata = []
```

```python
for i in archive.namelist():
    if i.endswith('.yaml'):
        yamldata.append(archive.read(i))
```
